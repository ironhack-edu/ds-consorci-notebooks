## If / Else en comprensi칩n

Las comprensiones de lista tambi칠n pueden incluir declaraciones condicionales con una estructura `if-else`. Esto te permite aplicar diferentes transformaciones o filtros a los elementos basados en condiciones espec칤ficas. **Lo que sabemos:**

````python
# Lista original de nombres de ciudades
list_of_words = ["Barcelona", "Madrid", "Girona", "Murcia"]

# Inicializar una lista vac칤a para almacenar nombres de ciudades modificados
new_list = []

# Iterar a trav칠s de cada nombre de ciudad en la lista original
for i in list_of_words:
    # Verificar si la longitud del nombre de la ciudad es mayor a 6 caracteres
    if len(i) > 6:
        # Si es as칤, convertir el nombre de la ciudad a may칰sculas y a침adirlo a new_list
        new_list.append(i.upper())
    else:
        # Si no, convertir el nombre de la ciudad a min칰sculas y a침adirlo a new_list
        new_list.append(i.lower())

# new_list contiene los nombres de ciudades modificados
new_list
````

En las comprensiones de listas anidadas, puedes tener una o m치s cl치usulas `for` e incluso incluir expresiones condicionales. Esta flexibilidad te permite construir estructuras de datos complejas y realizar operaciones avanzadas sobre datos anidados.

**Ejemplo:**

````python
nested_list = [[1, 2], [3,4], [5, 6]]
#output esperado, flat_list = [1, 2, 3, 4, 5, 6]
````

### For loop

Ahora creamos una lista de listas con el for loop

````python
new_list = []  # Crea una lista vac칤a para almacenar los elementos aplanados.

for i in nested_list:  # Itera a trav칠s de la lista exterior.
    for k in i:  # Itera a trav칠s de la lista interior dentro de la lista exterior.
        new_list.append(k)  # Agrega el elemento individual 'k' a new_list.

# El resultado ser치 una lista aplanada que contiene todos los elementos de la lista anidada.
````

### Comprehension list: nested lists

````python
flattened_list = [k for i in nested_list for k in i]
flattened_list
````

#### Desaplanando una lista anidada con una comprensi칩n de lista


````python
nested = [[[[1, 2], [3, 4]]]]
````

````python
new_list = []

for i in nested:
    for k in i:
        for j in k:
            for h in j:
                new_list.append(h)
                
new_list
````

````python
unnesting = [h for i in nested for k in i for j in k for h in j]
unnesting
````

### Sintaxis Regular vs. Comprensi칩n

Al trabajar con listas, tienes dos opciones principales: usar la sintaxis regular o usar comprensiones de lista. Cada enfoque tiene sus propias ventajas y casos de uso.

#### Sintaxis Regular

- **M치s F치cil de Escribir:** Los bucles regulares usando declaraciones for y if son m치s flexibles y te permiten tener un control completo sobre lo que est치s haciendo. Puedes a침adir declaraciones de impresi칩n dentro de los bucles para prop칩sitos de depuraci칩n.

- **M치s Control:** Puedes usar bucles regulares para escribir l칩gicas y condicionales m치s complejas.

- **Imprimir Dentro de los Bucles:** La depuraci칩n es m치s f치cil con bucles regulares ya que puedes insertar declaraciones de impresi칩n para ver resultados intermedios.

#### Comprensi칩n

- **M치s Eficiente:** Las comprensiones de lista son m치s eficientes y computacionalmente menos costosas, lo que las convierte en una buena elecci칩n para operaciones simples. Pueden mejorar el rendimiento del c칩digo.

- **M치s Simple:** Las comprensiones de lista simplifican el c칩digo reduciendo el n칰mero de pasos requeridos. No hay necesidad de crear una nueva lista y usar declaraciones de a침adir.

- **M치s R치pido:** Las comprensiones de lista son m치s r치pidas de escribir y a menudo resultan en c칩digo m치s conciso.

- **Vers치tiles:** Se pueden usar con listas, diccionarios, conjuntos y otros objetos iterables.

En resumen, las comprensiones de lista son una forma concisa y eficiente de crear listas, especialmente cuando la transformaci칩n de datos o filtrado es sencillo. Pueden mejorar tanto la legibilidad del c칩digo como el rendimiento en tales casos.

## Soluci칩n

````python
# List of feedback comments
feedback_comments = [
    "The service was excellent and I love the product!",
    "The quality of the item was great.",
    "Unfortunately, the package arrived damaged.",
    "I had a superb experience with your company.",
    "The product was disappointing and terrible.",
]

# List of positive and negative keywords
positive_words = ["excellent", "love", "great", "superb"]
negative_words = ["terrible", "disappointing", "damaged"]

# Extract positive keywords using list comprehension
positive_keywords = [word for comment in feedback_comments for word in comment.lower().split() if word in positive_words]

# Extract negative keywords using list comprehension
negative_keywords = [word for comment in feedback_comments for word in comment.lower().split() if word in negative_words]

# Display the extracted keywords
print("Positive Keywords:", positive_keywords)
print("Negative Keywords:", negative_keywords)
````
