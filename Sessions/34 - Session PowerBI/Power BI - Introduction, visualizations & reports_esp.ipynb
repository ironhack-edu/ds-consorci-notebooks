## 1. Introducción a Power BI <a class="anchor" id="chapter1"></a>
### 1.1 ¿Qué es Power BI y cómo funciona? <a class="anchor" id="section_1_1"></a>

Power BI es una serie de herramientas que trabajan conjuntamente y nos ayudan a colocar diferentes fuentes de datos de manera coherente y comprensible.

Power BI se divide en tres partes principales:

- Power BI desktop
- Power BI service (SaaS)
- Power BI Mobile

La importancia de estas tres partes es que te ayudan a crear, diseñar, compartir y consumir información relevante para los negocios de manera más efectiva.

#### Comparar Power BI Desktop y el servicio de Power BI

### Cómo funciona Power BI Desktop

* Conectarse a datos, incluyendo múltiples fuentes de datos.
* Modelar los datos con consultas que construyan modelos de datos perspicaces y convincentes.
* Usar los modelos de datos para crear visualizaciones e informes.
* Compartir los archivos de informes para que otros puedan aprovecharlos, construir sobre ellos y compartirlos. Puedes compartir archivos .pbix de Power BI Desktop como cualquier otro archivo, pero el método más convincente es subirlos al servicio de Power BI.

### ¿Qué ofrece?
- Creación de informes y análisis para nuestra organización de manera ágil.
- Interfaz simple e intuitiva.
- Aumento de la productividad al analizar datos, permitiéndonos anticipar situaciones atípicas o riesgosas.
- Contenido personalizado para la organización: comparte contenido con tu equipo o toda la organización.
- Todo el equipo sincronizado: toma decisiones rápidas y seguras con una sola vista de tu empresa. Una sola vista de tu empresa es clave para asegurarse de que todos están usando los datos correctos.
- Exportación a otras herramientas como Excel y PowerPoint.

## 3. Limpieza y transformación de datos <a class="anchor" id="chapter3"></a>

### 3.1 La importancia de limpiar y transformar los datos antes de analizarlos <a class="anchor" id="section_3_1"></a>

La importancia de limpiar y transformar los datos, ya la conoces, sin embargo, hay casos en los que hay personas que solo son analistas de Power BI, es decir, estas personas suben archivos de datos crudos a Power BI y hacen la limpieza y transformación en una tecnología llamada "Editor de Power Query" que ayuda a los analistas a transformar y limpiar los datos desde cero.

#### ¿Qué es Power Query?

Power Query es un motor de transformación y preparación de datos. Power Query viene con una interfaz gráfica para obtener datos de fuentes y un Editor de Power Query para aplicar transformaciones. Dado que el motor está disponible en muchos productos y servicios, el destino donde se almacenarán los datos depende de dónde se usó Power Query. Usando Power Query, puedes realizar el procesamiento de extracción, transformación y carga (ETL) de datos.



### 3.2 Cómo limpiar y transformar datos en Power BI usando el Editor de Consultas <a class="anchor" id="section_3_2"></a>
 
#### - Promocionar encabezados

Puedes promocionar encabezados de dos maneras: seleccionando la opción Usar primera fila como encabezados en la pestaña Inicio o seleccionando el botón desplegable junto a Columna1 y luego seleccionando Usar primera fila como encabezados.

#### - Renombrar columnas

Puedes renombrar los encabezados de las columnas de dos maneras. Un enfoque es hacer clic derecho en el encabezado, seleccionar Renombrar, editar el nombre y luego presionar Enter. Alternativamente, puedes hacer doble clic en el encabezado de la columna y sobrescribir el nombre con el nombre correcto.

#### - Eliminar filas superiores 

Selecciona Eliminar filas > Eliminar filas superiores en la pestaña Inicio.

#### - Eliminar columnas

Puedes eliminar columnas de dos maneras. El primer método es seleccionar las columnas que deseas eliminar y luego, en la pestaña Inicio, seleccionar Eliminar columnas.

#### Despivotar columnas

Despivotar es una característica útil de Power BI. Puedes usar esta característica con datos de cualquier fuente de datos, pero lo usarías más a menudo al importar datos desde Excel.

Selecciona la pestaña Transformar en Power Query y luego selecciona Despivotar.

#### Pivotar columnas

Puedes usar la característica de Columna Pivote para convertir tus datos planos en una tabla que contenga un valor agregado para cada valor único en una columna. Por ejemplo, podrías querer usar esta característica para resumir datos usando diferentes funciones matemáticas como Contar, Mínimo, Máximo, Mediano, Promedio o Suma.

En la pestaña Transformar, selecciona Transformar > Pivotar columnas

#### Reemplazar valores

Puedes usar la característica Reemplazar valores en el Editor de Power Query para reemplazar cualquier valor con otro valor en una columna seleccionada.

#### Eliminar duplicados

También puedes eliminar duplicados de columnas para mantener solo nombres únicos en una columna seleccionada usando la característica Eliminar Duplicados en Power Query.

Puedes lograr esta acción seleccionando una columna, haciendo clic derecho en el encabezado de la columna y luego seleccionando la opción Eliminar Duplicados.

Podrías considerar copiar la tabla antes de eliminar los duplicados. La opción de Copiar está en la parte superior del menú contextual, como se muestra en la siguiente captura de pantalla. Copiar la tabla antes de eliminar duplicados te dará una comparación de las tablas y te permitirá usar ambas tablas, si es necesario.

#### Cambiar el tipo de datos de la columna

En el Editor de Power Query, puedes cambiar el tipo de datos de la columna de dos maneras. Una manera es seleccionar la columna que tiene el problema, seleccionar Tipo de Datos en la pestaña Transformar y luego seleccionar el tipo de datos correcto de la lista.

Otro método es seleccionar el icono del tipo de datos junto al encabezado de la columna y luego seleccionar el tipo de datos correcto de la lista.

### 4.2 Cómo crear un panel de control y añadir visualizaciones y otros componentes <a class="anchor" id="section_4_2"></a>

Crear visualizaciones en Power BI es muy intuitivo. Para crear visualizaciones debemos ir a la vista de Informe y arrastrar a la página en blanco el tipo de visualización que queremos desde el panel de visualizaciones.

Existen muchos tipos de informes, ventas, tablas, recursos humanos, KPIs, financieros, etc.

Dependiendo del tipo de informe que quieras realizar, utilizarás una u otra visualización.

En el lado derecho encontrarás la sección de Visualizaciones, donde encontrarás las diferentes visualizaciones que puedes realizar. Luego, está la sección de Datos, donde puedes arrastrar las columnas, medidas y dimensiones que necesites, dependiendo de si necesitas que sean parte de la leyenda, el eje x o el eje y de un gráfico.

#### Descripciones emergentes (Tooltips)

Las descripciones emergentes son visualizaciones que nos ayudan a dar información más detallada sobre un gráfico. Pueden ser simples de 2 líneas o pueden contener visualizaciones en sí mismas.



## 5. Creando dashboards  <a class="anchor" id="chapter5"></a>

### 5.1 El propósito y los beneficios de crear un panel de control en Power BI <a class="anchor" id="section_5_1"></a>

Los paneles de control son importantes para mantener toda la información de una empresa consolidada en una visión más comprensiva. Ayudan a mantener varios tipos de gráficos de una manera relacionada y dinámica en un único informe.

Los paneles de control son una forma de monitorear tu negocio y ver todas tus métricas más importantes de un vistazo. Las visualizaciones en un panel pueden provenir de un conjunto de datos subyacente o de muchos, y de un informe subyacente o de muchos. Un panel combina datos locales y en la nube, proporcionando una vista consolidada de los datos.

Un panel no es solo una imagen bonita. Es altamente interactivo, y los mosaicos se actualizan a medida que los datos subyacentes cambian.

### 5.2 Cómo crear un panel de control y añadir visualizaciones y otros componentes <a class="anchor" id="section_5_2"></a>

Los paneles de control solo se pueden crear a través del servicio Power BI. Estos pueden ser visualizaciones ancladas (llamadas Baldosas) de informes ya creados, puedes enriquecer el panel desde varias fuentes de datos diferentes. Los paneles solo tienen una página.

## 6. Características avanzadas <a class="anchor" id="chapter6"></a>
    
### 6.1 Algunas de las características más avanzadas de Power BI, como medidas, fórmulas DAX y relaciones entre tablas <a class="anchor" id="section_4_2"></a>

#### - Medidas

Las medidas se utilizan en algunos de los análisis de datos más comunes. Resúmenes simples, como sumas, promedios, mínimo, máximo y conteos, también se pueden establecer a través del área de Campos. Los resultados calculados de las medidas están constantemente cambiando en respuesta a la interacción con los informes, permitiendo una exploración rápida y dinámica de datos ad hoc.

Las medidas de Power BI se crean a través del lenguaje DAX y su lenguaje de fórmulas. En DAX podemos encontrar funciones similares a las de Excel, como SUM y DATE, con la condición de que las funciones de DAX deben usarse con datos relacionales.
   
### - ¿Qué es DAX?

DAX es una colección de funciones, operadores y constantes que se pueden usar en una fórmula o expresión, para calcular y devolver uno o más valores. DAX te ayuda a crear nueva información a partir de datos existentes en el modelo.
