## 1. Introduction to Power BI <a class="anchor" id="chapter1"></a>

### 1.1 What is Power BI and how does it work? <a class="anchor" id="section_1_1"></a>

Power BI is a series of tools that work together and help us to place different data sources in a coherent and understandable way. 

Power BI is divided into three main parts:

- Power BI desktop
- Power BI service (SaaS)
- Power BI Mobile

The importance of these three parts is that they help you create, design, share and consume business relevant information in a more effective manner.

#### Compare Power BI Desktop and the Power BI service

### How Power BI Desktop works

* Connect to data, including multiple data sources.
* Shape the data with queries that build insightful, compelling data models.
* Use the data models to create visualizations and reports.
* Share your report files for others to leverage, build upon, and share. You can share Power BI Desktop .pbix files like any other files, but the most compelling method is to upload them to the Power BI service


### What does it provide?
- Creation of reports and analysis for our organization in an agile way.
- Simple and intuitive interface
- Increased productivity when analyzing data, allowing us to anticipate atypical or risky situations.
- Personalized content for the organization - share content to your team or the entire organization.
- The whole team in sync - Make quick and confident decisions with a single view of your company.A single view of your company. Key to making sure everyone is using the right everyone is using the right data.
- Export to other Excel and PowerPoint tools.

## 3. Data cleaning and transformation <a class="anchor" id="chapter3"></a>

### 3.1 The importance of cleaning and transforming data before analyzing it <a class="anchor" id="section_3_1"></a>

The importance of cleaning and transforming data, you already know, however, there are cases in which there are people who are only analysts from Power BI, that is, these people upload raw data files to Power BI and do the cleaning and transformation in a technology called "Power Query Editor" which helps analysts to transform and clean the data from scratch.


#### What is Power Query?

Power Query is a data transformation and data preparation engine. Power Query comes with a graphical interface for getting data from sources and a Power Query Editor for applying transformations. Because the engine is available in many products and services, the destination where the data will be stored depends on where Power Query was used. Using Power Query, you can perform the extract, transform, and load (ETL) processing of data.



 ### 3.2 How to clean and transform data in Power BI using the Query Editor <a class="anchor" id="section_3_2"></a>
 
#### - Promote headers

You can promote headers in two ways: by selecting the Use First Row as Headers option on the Home tab or by selecting the drop-down button next to Column1 and then selecting Use First Row as Headers.

#### - Rename columns

You can rename column headers in two ways. One approach is to right-click the header, select Rename, edit the name, and then press Enter. Alternatively, you candouble-click the column header and overwrite the name with the correct name.

#### - Remove top rows 

Select Remove Rows > Remove Top Rows on the Home tab.

#### - Remove columns

You can remove columns in two ways. The first method is to select the columns that you want to remove and then, on the Home tab, select Remove Columns.

#### Unpivot columns

Unpivoting is a useful feature of Power BI. You can use this feature with data from any data source, but you would most often use it when importing data from Excel.

Select the Transform tab in Power Query, and then select Unpivot.

#### Pivot columns

You can use the Pivot Column feature to convert your flat data into a table that contains an aggregate value for each unique value in a column. For example, you might want to use this feature to summarize data by using different math functions such as Count, Minimum, Maximum, Median, Average, or Sum.

On the Transform tab, selectTransform>Pivot Columns

#### Replace values

You can use the Replace Values feature in Power Query Editor to replace any value with another value in a selected column.

#### Remove duplicates

You can also remove duplicates from columns to only keep unique names in a selected column by using the Remove Duplicates feature in Power Query.

You can achieve this action by selecting a column, right-clicking on the header of the column, and then selecting the Remove Duplicates option.

You might consider copying the table before removing the duplicates. The Copy option is at the top of the context menu, as shown in the following screenshot. Copying the table before removing duplicates will give you a comparison of the tables and will let you use both tables, if needed.

#### Change the column data type

In Power Query Editor, you can change the column data type in two ways. One way is to select the column that has the issue, select Data Type in the Transform tab, and then select the correct data type from the list.

Another method is to select the data type icon next to the column header and then select the correct data type from the list.

## 5. Creating dashboards  <a class="anchor" id="chapter5"></a>

### 5.1 The purpose and benefits of creating a dashboard in Power BI <a class="anchor" id="section_5_1">

    Dashboards are important in order to keep all the information of a company consolidated in a more comprehensive overview. Dashboards help us to keep various types of graphs in a related and dynamic way in a single report.
    
    Dashboards are a way to monitor your business and see all your most important metrics at a glance. The visualizations on a dashboard can come from one underlying dataset or many, and from one underlying report or many. A dashboard combines on-premises and cloud data, providing a consolidated view of data.

    A dashboard isn't just a pretty picture. It's highly interactive, and the tiles update as the underlying data changes.




### 5.2 How to create a dashboard and add visualizations and other components <a class="anchor" id="section_5_2">

    Dashboards can only be created through Power BI service. These can be pinned visualizations (called Tiles) from already created reports, you can enrich the dashboard from several different data sources. Dashboards only have a single page.
    

## 6. Advanced features <a class="anchor" id="chapter6"></a>

### 6.1 Some of the more advanced features of Power BI, such as measures, DAX formulas, and relationships between tables <a class="anchor" id="section_4_2"></a>

#### - Measures

   Measures are used in some of the most common data analyses. Simple summaries, such as sums, averages, minimum, maximum and counts, can also be set through the Fields area. The calculated results of the measures are constantly changing in response to interaction with reports, allowing for quick and dynamic exploration of ad hoc data.

   Power BI measures are created through the DAX language and its formula language. In DAX we can find functions similar to those of Excel, such as SUM and DATE, with the condition that DAX functions must be used with relational data.
   
### - What is DAX?

DAX is a collection of functions, operators and constants that can be used in a formula or expression, to calculate and return one or more values. DAX helps you create new information from existing data in the model.
